```python
import pandas as pd
```

```python
world_data1= pd.read_csv("world-data-2023_part1.csv", index_col=0)
world_data1
```

```python
world_data2= pd.read_csv("world-data-2023_part2.csv", index_col=0)
world_data2
```

```python
world_data2.columns
```

```python
world_data1.columns
```

columnas coincidentes:
- Country en WD1
- country en WD2

```python
# union de df
world_data_unido = world_data1.merge(world_data2, left_on="Country", right_on="country", how="left")
world_data_unido
```

```python
#nos cargamos country

world_data_unido.drop(columns=["country"], inplace=True)
world_data_unido
```

Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

- No tengan espacios
- quitar \n
- Esten en .lower()
- quitar (:)

```python
world_data_unido.dtypes
```

```python
world_data_unido[world_data_unido.select_dtypes(include='object').columns] = world_data_unido.select_dtypes(include='object').apply(lambda x: x.str.strip())
world_data_unido
```

```python
world_data_unido.columns
```

```python
# Quitar los paréntesis y el contenido dentro de ellos sin usar regex
world_data_unido.columns = [col.split('(')[0] for col in world_data_unido.columns]
world_data_unido.columns
```

```python
#quitamos \n

world_data_unido.columns = world_data_unido.columns.str.replace("\n", "")
world_data_unido.columns
```

```python
# quitamos (:)

world_data_unido.columns = world_data_unido.columns.str.replace(":", "")
world_data_unido.columns
```

```python
# ponemos en micuscula

world_data_unido.columns = world_data_unido.columns.str.lower()
world_data_unido.head()
```


La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
world_data_unido["coordinates"].str.split(",")  # se hace spñit por , porque es el elemento que separa los numeros en la columna
```

```python
world_data_unido["coordinates"].str.split(",", expand = True)

```

```python
world_data_unido[["longitud", "latitud"]] = world_data_unido["coordinates"].str.split(",", expand = True)

# comprobamos que se han creado correctamente las columnas
world_data_unido.head(2)
```

```python
world_data_unido.dtypes
```

```python
# borramos "coordinates"

world_data_unido.drop(columns=['coordinates'], inplace=True)

world_data_unido
```
