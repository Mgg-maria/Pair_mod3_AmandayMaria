```python
# Manera 1: Bucle sin utilizar apply

import pandas as pd

def clean_numeric_columns(df, columns):
    """
    Limpia y convierte las columnas especificadas de un DataFrame a enteros,
    eliminando comas y manejando valores NaN.
    
    :param df: DataFrame de Pandas
    :param columns: Lista de columnas a limpiar y convertir
    :return: DataFrame con columnas convertidas
    """
    for col in columns:
        df[col] = (
            df[col]
            .astype(str)  # Asegurar que todos los valores sean cadenas
            .str.replace(",", "", regex=True)  # Eliminar comas en números
            .replace("nan", None)  # Reemplazar texto "nan" con None
            .astype("Int64")  # Convertir a enteros permitiendo NaN
        )
    return df
```

```python
# Manera 2: Utilizando apply para aplicarlo en todas las columnas

import pandas as pd

def clean_numeric_column(col):
    """
    Limpia y convierte una columna a enteros, eliminando comas y manejando NaN.
    """
    return (
        col.astype(str)  # Convertir a string
        .str.replace(",", "", regex=True)  # Eliminar comas
        .replace(["nan", "<NA>", "None", ""], None)  # Reemplazar texto "nan" por None
        .astype(float)  # Convertir a enteros permitiendo NaN
    )

# Aplicar la función a múltiples columnas con apply
columns_to_convert = ["land_area","armed_forces_size", "co2-emissions", "gdp", "population","urban_population","latitude","longitude","density","cpi"]  # Reemplaza con tus columnas
world_data_full[columns_to_convert] = world_data_full[columns_to_convert].apply(clean_numeric_column)
```

```python
# Creamos una función que nos da el país y el continente al que pertenece. 
# Este código nos da el país pero NO puede indicarnos a qué continente pertenece. Para ello debemos indicárselo (ver siguiente celda de código)

from geopy.geocoders import Nominatim    # Esto nos permite convertir nombres de lugares en coordenadas o extraer información de una ubicación.

def obtener_pais_continente(ubicacion):  # Se define la función obtener_pais_continente(ubicacion), que toma como entrada un nombre de ciudad, país o dirección.
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(ubicacion, exactly_one=True, addressdetails=True)

    if location and "address" in location.raw:
        pais = location.raw["address"].get("country", "Desconocido")
        continente = location.raw["address"].get("continent", "Desconocido")
        return pais, continente
    else:
        return "No encontrado", "No encontrado"

# Ejemplo de uso
pais, continente = obtener_pais_continente("Madrid, Spain")
print(f"País: {pais}, Continente: {continente}")
```
